```python
# Melakukan import packages pandas sebagai pd
import pandas as pd
```

```python
# Melakukan import dataset dan menampilkan 5 data teratas dari dataset superstore.csv
df = pd.read_csv('superstore.csv')
df.head()
```

```python
# Mengelompokkan data berdasarkan kategori produk dan menghitung total penjualan dan keuntungan
grouped_data = df.groupby('Category').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Menampilkan hasil pengelompokan
print(grouped_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat visualisasi data bar plot untuk penjualan dan keuntungan berdasarkan kategori
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Sales', data=grouped_data, palette='viridis')
plt.title('Total Penjualan Berdasarkan Kategori')
plt.xlabel('Kategori')
plt.ylabel('Total Penjualan')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Profit', data=grouped_data, palette='viridis')
plt.title('Total Keuntungan Berdasarkan Kategori')
plt.xlabel('Kategori')
plt.ylabel('Total Keuntungan')
plt.show()
```
